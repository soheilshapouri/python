```python
from scipy import stats
```

```python
cor_coef, p = stats.pearsonr(df2['price'],df2['horsepower'])
print(cor_coef,p)
```

```python
Cor_coef, p_value= stats.pearsonr(df2['price'], df2['length'])
print(f"P-value: {p_value:.2f}")
```

```python
#one-way ANOVA
df2['drive-wheels'].unique()
#compare the prices of three drive wheels with one way ANOAV
```

```python
rwd = df2[df2['drive-wheels'] == 'rwd']['price']
fwd = df2[df2['drive-wheels'] == 'fwd']['price']
fourwd = df2[df2['drive-wheels'] == '4wd']['price']
```

```python
stats.f_oneway(rwd, fwd, fourwd)
```

```python
#similar to ad hoc tests but probably wrong way to do it 
stats.f_oneway(fwd, rwd)
```
