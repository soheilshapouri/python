We can create a dataframe from a dictionary. THe keys of the dictionary corresponde to columns. 
the values of te dictionary are rows. 
Finally we can cast the dictionary to a dataframe by using DataFrame() from pandas. 


```python
SampleDisasters = {"Disasters":["Coal mine", "Santa Maria", "Mount Karthala"], "Country": ["Belgium", "Guatemala", "Comoros"], 
                   "Year":[1901, 1902, 1903]}
type(SampleDisasters)
```

```python
SampleDisasters_df = pd.DataFrame(SampleDisasters)
print(type(SampleDisasters_df))
SampleDisasters_df.head()
```

```python
# we can make a dataframe from one column of another dataframe
x = SampleDisasters_df[["Disasters"]]
type(x)
```
