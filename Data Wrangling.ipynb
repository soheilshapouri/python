Data Wrangling: changing the format of the data so it is better for analysis 

```python
import pandas as pd
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/autos/imports-85.data"
```

```python

df = pd.read_csv(url, header= None)
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
df.columns = headers

#instead of using df.columns = headers we can use this code df = pd.read_csv(filename, names = headers)
```

```python
#because price is the outcome the best option for dealing with missing values of price is to remove the entire row 
df.dropna(subset=["price"], axis = 0, inplace= True)
#axis = 0 drop the entire row
#axis = 1 drop the entire column 
```

```python
df.describe(include ="all")
```

```python
#becuase nan is treated as srt we use astype('float'); can ignore axis = 0 
#using mean instead of missing values here 
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
# find the mode
df['num-of-doors'].value_counts().idxmax()
```

```python
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)

df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
#change the data format 
df[["bore", "stroke"]] = df[["bore", "stroke"]].astype("float")
df[["normalized-losses"]] = df[["normalized-losses"]].astype("int")
df[["price"]] = df[["price"]].astype("float")
df[["peak-rpm"]] = df[["peak-rpm"]].astype("float")
```

```python
#after droping rows with missing values it is better to reset index 
# simply drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
# there are some missing values for normalized losses so let's relapce them with the mean of that column 
mean = df['normalized-losses'].mean()
```

```python
df['normalized-losses']=pd.to_numeric(df['normalized-losses'], errors='coerce')
```

```python
mean = df['normalized-losses'].mean()
df['normalized-losses'].replace(np.nan,mean)
```

```python
df['price']= df['price'].astype("int")
```

Data Normalization 

simple feature scaling: xnew = xold / xmax --> gives 0 to 1

min-max normalization: xnew = (xold - x min) / (xmax - xmin) --> fives 0 to 1 

z score: xnew = xold - mu / sigma --> typicaly score will range from -3 and +3 

```python
# simple feature scaling 
df['variable']= df['variable'] / df['variable'].max()
```

```python
# min-max normalization
df['variable'] = (df['variable'] - df['variable'].min()) / (df['variable'].max() - df['variable'].min())
```

```python
# z-score
df['variable'] = (df['variable'] - df['variable'].mean()) / df['variable'].std()
```
