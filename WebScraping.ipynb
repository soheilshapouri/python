We need requests and Beautiful Soup modules to scrape websites 

```python
from bs4 import BeautifulSoup  #beautiful soup is used for extarcting data from HTML files 
import requests
import pandas as pd
```

```python
page = requests.get('https://en.wikipedia.org/wiki/Disaster').text

#now page is a very big string object

```

```python
#create a Beautiful soup object from the variable page 
 
soup = BeautifulSoup(page, "html.parser")
soup
#this soup object will make data like nested 
```

```python
print(soup.prettify())
```

```python
#now the soup object has attributs and we can use beautifulsoup methods on it 
soup.title
```

```python
#to find a specific type of title 
h3tag = soup.h3
h3tag
```

```python
#find the hyperlinks in the html 
soup.find_all(href = True)
```

beautiful soup documentation has a lot of useful easy to uderstand material to work with html files. 


```python
#find the links in the html page 
for link in soup.find_all('a',href=True):  # in html anchor/link is represented by the tag <a>

    print(link.get('href'))
```

```python
# Let's find all the tables in the html. remember table is a html tag
tables = soup.find_all("table")
```

```python
#there should be one table 
len(tables)
#probably beacuse many things that don't look like a table ate actually formatted as a table 
```

```python
#find the table called Disaster
for index,table in enumerate(tables):
    if ("Disaster" in str(table)):
        table_index = index
print(table_index)
```
