```python
#now I create a client object 
cg = CoinGeckoAPI()
```

```python
bitcoin_data = cg.get_coin_market_chart_by_id(id = 'bitcoin', vs_currency = 'usd', days = 30)
```

```python
#bitcoin data now has three variables but I am only interested in prices 
bitcoin_price_data = bitcoin_data['prices']
```

```python
#change the dictionary to a dataframe 
import pandas as pd
```

```python
data = pd.DataFrame(bitcoin_price_data, columns = ['TimeStamp','Price'])
data
```

```python
# it is difficult to read the time stamps above so let's change them 
data['Date'] = pd.to_datetime(data['TimeStamp'], unit= 'ms')
data
```

```python
#now we want to create a candleStick plots as they are common in financial analysis 
#but first we should change the format of teh data to have min, max, first, and last data points for each date 

candlestick_data = data.groupby(data.Date.dt.date).agg({'Price': ['min', 'max', 'first', 'last']})
candlestick_data
```
